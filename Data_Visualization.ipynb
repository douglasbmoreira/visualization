#Biblioteca Matplotlib

## Unidade 1 - Conhecendo a biblioteca Matplotlib

### Aula 1 - Conhecendo os dados

```python
#importando a biblioteca pandas e lendo o arquivo csv
import pandas as pd
df = pd.read_csv('/content/imigrantes_canada.csv')
df.head()
```

```python
#verificando a quantidade de linhas e colunas da planilha
df.shape
```

```python
#verificando os tipos de dados de cada coluna
df.info()
```

```python
#Transformando a coluna 'País' em index
df.set_index('País', inplace= True)
df
```

```python
#Criando uma variável, alocando os anos que temos na planilha dentro de uma lista em formato de string
anos = list(map(str, range(1980, 2014)))
anos
```

```python
#Criando uma variável com as informações do Brasil, localizando os dados com a função .loc
brasil = df.loc['Brasil', anos]
brasil
```

```python
#Criando um dicionário com os dados do Brasil
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes':brasil.values.tolist()}
brasil_dict
```

```python
#Criando um Dataframe com os dados do dicionário
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

### Aula 2 - Criando gráfico no Matplotlib

```python
#Importando a biblioteca Matplotlib usada em visualização dos dados
import matplotlib.pyplot as plt
```

```python
#Plotando um gráfico de linha dos dados do dataframe dados_brasil
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

### Aula 3 - Adicionando títulos e legendas

```python
#Adicionando legendas para os eixos do gráfico
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Anos')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

### Aula 4 - Criando subplots em duas direções parte 1

### Aula 5 - Criando subplots em duas direções parte 2

### Aula 3 - Alterando as cores com paletas

```python
#Criando uma função com a função def para facilitar a criação de novos gráficos
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', hue=top_10.index, palette=palette)
  ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=16)
  ax.set_xlabel('Número de imigrantes', fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

### Para saber mais: Animando gráficos para mostrar mudanças ao longo do tempo

```python
#Alterando o tipo de dados dos anos de string para inteiro
dados_brasil['ano']= dados_brasil['ano'].astype(int)
```

```python
import plotly.graph_objs as go

# Criando uma figura
fig = go.Figure()

# Adicionando a linha do gráfico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]], mode='lines', name='Imigrantes', line=dict(width=4))
)

# Definindo as configurações de layout
fig.update_layout(
    title=dict(
        text='<b>Imigração do Brasil para o Canadá no período de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Definir as configurações de animação
frames = [go.Frame(data=[go.Scatter(x=dados_brasil['ano'].iloc[:i+1], y=dados_brasil['imigrantes'].iloc[:i+1])]) for i in range(len(dados_brasil))]
fig.frames = frames

# Mostrando a figura
fig.show()
```
